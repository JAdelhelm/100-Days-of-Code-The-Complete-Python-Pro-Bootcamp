```python
import pandas as pd

df = pd.read_csv('salaries_by_college_major.csv')
df.head()
```

```python
diff_accounting = df[df["Undergraduate Major"] == "Accounting"]["Mid-Career 90th Percentile Salary"] - df[df["Undergraduate Major"] == "Accounting"]["Mid-Career 10th Percentile Salary"]
diff_accounting
```

```python
df_acc = df[df["Undergraduate Major"] == "Accounting"]
df_acc['Mid-Career 90th Percentile Salary'].subtract(df_acc['Mid-Career 10th Percentile Salary'])
```

```python
## Add differences as a new column

spread_col = df["Mid-Career 90th Percentile Salary"] - df["Mid-Career 10th Percentile Salary"]
spread_col
```

```python
df["Spread"] = spread_col
df.head(3)
```

```python
df.sort_values("Spread", ascending=True).head(3)
```
